```python
# Ejercicio 1

email = "usuario@dominio.com"

def get_domain(email:str) -> str:
  pos_arr = email.index('@')
  # pos_p = email.index('.')
  return email[pos_arr+1::]

print(get_domain(email=email))
```

```python
# Ejercicio 2

def identify_sell_level(is_pastas: int, cant:int)->None:
  ans = ["Venta Baja", "Venta Regular", "Venta Grande"]
  
  limits = {
    'baja':[0, 60],
    'regular':[61, 80],
    'grande':81
  }

  if is_pastas == 0 or is_pastas == 1:
    if is_pastas == 1:
      limits['baja'][1] += 50
      limits['regular'][1] += 50
      limits['grande'] += 50 

    if cant >= limits['baja'][0] and cant <= limits['baja'][1]:
      return ans[0]
    elif cant >= limits['regular'][0] and cant <= limits['regular'][1]:
      return ans[1]
    elif cant >= limits['grande']:
      return ans[2]
  else:
    return "Alert! invalid pastas value..."

print(identify_sell_level(is_pastas=0, cant=50))
print(identify_sell_level(is_pastas=0, cant=70))
print(identify_sell_level(is_pastas=0, cant=90))

print(identify_sell_level(is_pastas=1, cant=50))
print(identify_sell_level(is_pastas=1, cant=120))
print(identify_sell_level(is_pastas=1, cant=140))

print(identify_sell_level(is_pastas=2, cant=140))
```

```python
# Ejercicio 3
import pandas as pd

df = pd.read_csv('Prueba_Moderno.csv', sep=';', encoding = "ISO-8859-1")
df
```

```python
import matplotlib.pyplot as plt
by_Cadena = df.value_counts('Cadena').reset_index()
plt.figure(figsize=(5, 5))
plt.title('Porcentaje de ventas por Cadena')
plt.grid(visible=True)
plt.pie(
  by_Cadena['count'], 
  labels = by_Cadena['Cadena'], 
  autopct='%1.1f%%',
)
plt.show()
```

```python
import matplotlib.pyplot as plt
by_Categoria = df.value_counts('Categoría').reset_index()
fig, ax = plt.subplots()


hbars = ax.barh(by_Categoria['Categoría'],by_Categoria['count'])
for bar in hbars:
  bar.set_color('blue')
ax.set_ylabel('Categorias')
ax.set_title('Cantidad de ventas por categorías')
ax.bar_label(hbars)
plt.show()
```
